# data visualization of Tourisam project

# STEP 1: Load Cleaned Dataset


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_tourism_data.csv")
df.head()
```

STEP 2: Check Rating Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Rating"], bins=5, kde=True)
plt.title("Rating Distribution")
plt.show()
```

STEP 3: Visit Mode Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="VisitMode")
plt.xticks(rotation=45)
plt.title("Visit Mode Distribution")
plt.show()
```

STEP 4: Top Attraction Types

```python
top_types = df["AttractionType"].value_counts().head(10)
top_types.plot(kind="bar")
plt.title("Top 10 Attraction Types")
plt.show()
```

STEP 5: Average Rating per Visit Mode

```python
df.groupby("VisitMode")["Rating"].mean().plot(kind="bar")
plt.title("Average Rating by Visit Mode")
plt.show()
```
