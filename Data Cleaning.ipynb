Fix Rating Column

```python
transaction["Rating"] = pd.to_numeric(transaction["Rating"], errors="coerce")
transaction = transaction.dropna(subset=["Rating"])
```

MERGE LOCATION DATA
# Merge User + City

```python
user_city = user.merge(city, on="CityId", how="left")
```

# Merge Country, Region, Continent

```python
user_location = (
    user_city
    .merge(country, left_on="CountryId_x", right_on="CountryId", how="left")
    .merge(region, left_on="RegionId_y", right_on="RegionId", how="left")
    .merge(continent, left_on="ContinentId_y", right_on="ContinentId", how="left")
)



```

# MERGE ATTRACTION DATA

```python
item_full = item.merge(atype, on="AttractionTypeId", how="left")
```

# FINAL MASTER MERGE

```python
final_df = (
    transaction
    .merge(user_location, on="UserId", how="left")
    .merge(item_full, on="AttractionId", how="left")
)
```

# Remove Duplicates

```python
final_df.drop_duplicates(inplace=True)
```
