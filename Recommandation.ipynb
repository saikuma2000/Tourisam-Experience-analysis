# Recommendation System (Simple & Clean)

STEP 1: Create User-Item Matrix

```python
import pandas as pd

df = pd.read_csv("dataset/final_model_data.csv")

print(df.head())
user_item = df.pivot_table(
    index="UserId",
    columns="AttractionId",
    values="Rating"
)

print(user_item.head())
```

STEP 2: Correlation-Based Recommendation

```python
def recommend(user_id, top_n=5):
    user_ratings = user_item.loc[user_id]
    similar_scores = user_item.corrwith(user_ratings)
    similar_scores = similar_scores.sort_values(ascending=False)
    return similar_scores.head(top_n).index.tolist()
```
